#### Data Preparation:

Defining the XOR dataset as two NumPy arrays: 

- X represents the input features (two binary values) for each data point, and 
- y represents the corresponding desired output (0 or 1).

```python
# Defining the XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # input 
y = np.array([[0], [1], [1], [0]])    # output
```

#### Hyperparameter Configuration:

- input_size: The number of input features (2 in this case).
- hidden_size: The number of neurons in the hidden layer. using 2 for simplicity
- output_size: The number of output neurons (1 for binary classification).
- lr: The learning rate, which controls the magnitude of weight updates during training.
- epochs: The number of training iterations (epochs) the network will go through.

```python
input_size = 2
hidden_size = 2
output_size = 1
lr = 0.1
epochs = 10000
```

#### Weight and Bias Initialization:

- W_hidden: Randomly initialized weights between the input layer and the hidden layer using the np.random.randn function, which draws samples from a standard normal distribution.
- b_hidden: Randomly initialized biases for the hidden layer neurons.
- W_output: Randomly initialized weights between the hidden layer and the output layer.
- b_output: Randomly initialized biases for the output layer neuron.

```python

# intializing weights and biases with random values 

W_hidden = np.random.randn(input_size, hidden_size)  # range - inf to +inf  # normal distribution 
b_hidden = np.random.randn(1, hidden_size)

W_output = np.random.randn(hidden_size, output_size)
b_output = np.random.randn(1, output_size)


# we can intialize using uniform distribution aswell
# W1 = 2 * np.random.random((input_size,hidden_size)) -1  # range of values [-1,1]   , np.random.random() - uniform distribution - 0 to1
# b1 = 2 * np.random.random((1,hidden_size)) -1

print('Intial Hidden layer weights "W_hidden": \n',W_hidden)
print('\n Initial Hidden layer Biases "b_hidden": \n ',b_hidden)

print('\n\n Intial output layer weights "W_output": \n ',W_output )
print('\n Intial output layer Biases "b_output" :\n',b_output)
```

#### Sigmoid Activation Function (sigmoid(z))

- The sigmoid function, also known as the logistic function, is a commonly used activation function in neural networks.
- It takes a real-valued input z and squashes it between 0 and 1, introducing non-linearity into the network.
- The mathematical formula for the sigmoid function is 1 / (1 + exp(-z)), where exp is the exponential function.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

#### Derivative of Sigmoid Function (derivativeSigmoid(z))

- In backpropagation, we need to calculate the gradients (changes) of the loss function with respect to the weights and biases.
- This derivative of the sigmoid function is used during backpropagation to update the network weights and biases based on the error.
- The mathematical formula for the derivative of the sigmoid function is sigmoid(z) * (1 - sigmoid(z)). This utilizes a clever property of the sigmoid function where its derivative can be expressed in terms of itself.

```python
def derivativeSigmoid(z):
    return sigmoid(z) * (1-sigmoid(z))
```

This section defines the core training loop for our XOR neural network. It iterates through a specified number of epochs (epochs) to perform training:

#### Forward Propagation:

- *Input Layer*: The training data (X) is fed into the network.
- *Hidden Layer*:
    - hidden_layer_input : A weighted sum of the inputs and the hidden layer bias (b_hidden) is calculated for each neuron in the hidden layer.
    - hidden_layer_output : The sigmoid activation function (sigmoid) is applied to the hidden layer input, resulting in the hidden layer activation. This introduces non-linearity into the network's predictions.
- *Output Layer*:
    - output_layer_input : A weighted sum of the hidden layer activations and the output layer bias (b_output) is calculated.
    - output_layer_output: The sigmoid activation function is again applied to the output layer input, resulting in the network's prediction for the current training example.

#### Backpropagation:

- first_term_output_layer : The derivative of the error is calculated between the predicted output (output_layer_output) and the desired target output (y).
- second_term_output_layer : The derivative of the sigmoid function (derivativeSigmoid) is used to account for the non-linearity introduced by the activation function. It's multiplied (chain rule) by the error term for the output layer.
- Similar calculations are performed for the hidden layer (first_term_hidden_layer, second_term_hidden_layer, firstTwoTerms_Hidden_layer).

#### Weight and Bias Updates:

- Gradients (changes) for the weights (delta_output_weights, delta_hidden_weights) and biases (delta_output_bias, delta_hidden_bias) are computed using the error terms, activations, and the learning rate (lr). 
- These gradients indicate how much to adjust the weights and biases to minimize the error for future training examples.
- The weights and biases are updated proportionally to the learning rate and the calculated gradients. This iterative process allows the network to learn from the training data.

```python
# Training
for epoch  in range (epochs):
    
    # Forward propogation
    input = X
    hidden_layer_input = np.dot(input, W_hidden) + b_hidden
    hidden_layer_output = sigmoid(hidden_layer_input)
    output_layer_input = np.dot(hidden_layer_output, W_output) + b_output
    output_layer_output = sigmoid(output_layer_input)

    # Back propogation
    first_term_output_layer = output_layer_output - y 
    second_term_output_layer = derivativeSigmoid(output_layer_input)
    first_two_terms_output_layer = first_term_output_layer * second_term_output_layer

    first_term_hidden_layer = np.dot(first_two_terms_output_layer,W_output.T)
    second_term_hidden_layer = derivativeSigmoid(hidden_layer_input)
    firstTwoTerms_Hidden_layer = first_term_hidden_layer * second_term_hidden_layer

    delta_output_weights = np.dot(hidden_layer_output.T,first_two_terms_output_layer)
    delta_output_bias = np.sum(first_two_terms_output_layer, axis=0, keepdims=True)

    delta_hidden_weights = np.dot(input.T,firstTwoTerms_Hidden_layer)
    delta_hidden_bias = np.sum(firstTwoTerms_Hidden_layer,axis=0,keepdims=True)

    # weights and bias updations

    W_output = W_output - lr * delta_output_weights
    b_output = b_output - lr * delta_output_bias

    W_hidden = W_hidden - lr * delta_hidden_weights
    b_hidden = b_hidden - lr * delta_hidden_bias
```

```python
input = X
hidden_layer_input = np.dot(input, W_hidden) + b_hidden
hidden_layer_output = sigmoid(hidden_layer_input)
output_layer_input = np.dot(hidden_layer_output, W_output) + b_output
output_layer_output = sigmoid(output_layer_input)

print('Predicted_output: \n',output_layer_output)

print('\n Trained Hidden layer weights "W_hidden": \n',W_hidden)
print('Trained Hidden layer Biases "b_hidden": \n ',b_hidden)

print('\n Trained output layer weights "W_output": \n ',W_output )
print('Trained output layer Biases "b_output" :\n',b_output)
```
